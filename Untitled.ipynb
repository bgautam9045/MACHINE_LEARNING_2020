```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('Missing data 28-08-2020.csv')
data
```

```python
X = data.iloc[:, 0:3]
Y = data.iloc[:, 3]
```

```python
X = pd.DataFrame(X, columns = ['Country', 'Age', 'Salary'])
X
```

```python
Y = pd.DataFrame(Y, columns = ['Purchased'])
Y
```

```python
imp = SimpleImputer(missing_values = np.nan,strategy ='mean')
imp = imp.fit_transform(X.iloc[:, 1:3])
imp
```

```python
X.iloc[:, 1:3] = imp
X['Age'] = X['Age'].apply(lambda x : int(x))
X['Salary'] = X['Salary'].apply(lambda x : int(x))
X
```

```python
Y
```

```python
imp = SimpleImputer(missing_values = np.nan,strategy ='median')
print(imp.fit_transform(X.iloc[:, 1:3]))
```

```python
imp = SimpleImputer(missing_values = np.nan,strategy ='most_frequent')
print(imp.fit_transform(X.iloc[:, 1:3]))
```

```python
imp.fit(X.iloc[:, 1:3])
```

```python
print(imp.transform(X.iloc[:, 1:3]))
```
